```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
```

```python
classNames = { 0: 'background',
1: 'aeroplane', 2: 'bicycle', 3: 'bird',
4: 'boat', 5: 'bottle', 6: 'bus',
7: 'car', 8: 'cat', 9: 'chair',
10: 'cow', 11: 'diningtable', 12: 'dog',
13: 'horse', 14: 'motorbike', 15: 'person', 16: 'pottedplant',
17: 'sheep', 18: 'sofa', 19: 'train', 20: 'tvmonitor' }
```

```python
proto="/content/SSD_using_OpenCV/MobileNetSSD_deploy.prototxt"
weights="/content/SSD_using_OpenCV/MobileNetSSD_deploy.caffemodel"
net=cv2.dnn.readNetFromCaffe(proto,weights)
```

```python
img=cv2.imread("/content/SSD_using_OpenCV/doggo.jpg")
img_resized=cv2.resize(img,(300,300))
```

```python
blob = cv2.dnn.blobFromImage(img_resized , 0.007843 , (300 , 300)
 , (127.5 , 127.5 , 127.5) , False)
```

```python
net.setInput(blob)

```

```python
detections = net.forward()
```

```python
# prompt: detect the shape

detections.shape
```

```python
final=detections.squeeze()
```

```python
final
```

```python
height , width , _ = img.shape
```

```python
font=cv2.FONT_HERSHEY_COMPLEX
```

```python
ig=cv2.imread("/content/SSD_using_OpenCV/doggo.jpg")
for i in range(final.shape[0]):
   conf = final[i , 2]
   if conf > 0.01:
      class_name = classNames[final[i , 1]]
      x1n , y1n , x2n , y2n = final[i , 3:]
      x1 =int(x1n * width)
      y1 = int(y1n *height)
      x2 = int(x2n* width)
      y2 = int(y2n * height)
      top_left = (x1,y1)
      bottom_right = (x2 , y2)
      ig = cv2.rectangle(ig , top_left , bottom_right , (0 , 255 , 0) , 3)
      ig = cv2.putText(ig , class_name ,(x1 ,y1 - 10) , font , 0.5 , (255 , 0 , 0) ,
                       2 , cv2.LINE_AA)
```
